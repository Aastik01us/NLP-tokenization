```python
from serpapi import GoogleSearch
import pandas as pd
import time

# Load your company name file
df = pd.read_csv("/content/HR info 3.csv")

# Add an empty LinkedIn column
df["LinkedIn URL"] = ""

# üîê Paste your API key here
SERP_API_KEY = "fca69677467a439ef24cfb7ecb1ce50297ea26870a7b3b34b297c42b15d9bebc"

def get_linkedin_url(company_name):
    try:
        search = GoogleSearch({
            "q": f"{company_name} site:linkedin.com/company",
            "api_key": SERP_API_KEY,
            "num": 1
        })
        results = search.get_dict()
        if "organic_results" in results:
            for res in results["organic_results"]:
                link = res.get("link", "")
                if "linkedin.com/company" in link:
                    return link
    except Exception as e:
        print(f"Error for {company_name}: {e}")
    return "Not Found"
```

```python
# Loop through companies and fill LinkedIn URLs
for idx, row in df.iterrows():
    company = row["Company Name"]
    print(f"Searching: {company}")
    linkedin_url = get_linkedin_url(company)
    df.at[idx, "LinkedIn URL"] = linkedin_url
    time.sleep(2)  # SerpAPI rate limit safety
```

```python
# Save output
output_file = "HR_with_real_linkedin_urls.csv"
df.to_csv(output_file, index=False)

from google.colab import files
files.download(output_file)
```
